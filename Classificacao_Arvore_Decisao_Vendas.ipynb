```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel('/content/drive/MyDrive/vendas.xlsx')
dados.head()
```

```python
x = pd.get_dummies(dados[['media_rating','media_pedidos','test']])
x.head()
```

```python
y = dados['cliente_comprou']
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
modelo = arvore.fit(x,y)
modelo
```

```python
fig = plt.figure(figsize=(70,30))
fig = tree.plot_tree(modelo, feature_names=x.columns, class_names = ['Yes','no'], filled=True)
```
